# 3.Represent Corelation between variables ?

```python
corelation_variable = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corelation_variable)
plt.show()
```

# plot different kinds of graphs for mean value represetation ? (use required headers) ?

```python
mean_values= df.mean()
plt.figure(figsize=(10,8))
sns.boxplot(x="cylinders",y="weight",data=df)
plt.show()
```

```python
mean_values= df.mean()
plt.figure(figsize=(10,8))
sns.violinplot(x="cylinders",y="weight",data=df)
plt.show()
```

# Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4 ?

```python
sns.pairplot(df[["mpg","weight","origin"]],hue="origin",height=4)
plt.show()
```

# Identify how many types of  cylinder vehicles you have ? (GroupBY)

```python
cylinders=df.groupby('cylinders').value_counts()
print(cylinders)
```

# how many vehicles in every  model_year ?


```python

vehicle_counts_by_year = df.groupby('model_year').size().reset_index()

print(vehicle_counts_by_year)

```
